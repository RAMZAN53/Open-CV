```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
path=('/content/drive/MyDrive/OpenCV/Images/Earth.jpeg')
img=cv2.imread(path)
re_img=cv2.resize(img,(300,300))
cv2_imshow(re_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
path=('/content/drive/MyDrive/OpenCV/Images/Earth.jpeg')
img=cv2.imread(path)
re_img=cv2.resize(img,(300,300))
h=np.hstack((re_img,re_img,re_img,re_img))
v=np.vstack((h,h))
cv2_imshow(v)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import os
list=os.listdir('/content/drive/MyDrive/OpenCV/Images/') # Get a list of filenames in the directory
for name in list:
  path='/content/drive/MyDrive/OpenCV/Images/'
  img_name=os.path.join(path, name)
  img=cv2.imread(img_name)
  if img is not None:
    img=cv2.resize(img,(300,300))
    cv2_imshow(img)
    cv2.waitKey(0)
  else:
    print(f"Failed to load image: {img_name}")
cv2.destroyAllWindows()
```
